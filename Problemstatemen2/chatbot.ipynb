```python
def greenting_response(text):
    text=text.lower()
    
    bot_greetings=['howdy','hi','hey','hello','hola']
    user_greetings=['hi','hey','hola','hello','whatsup']
    
    
    for word in text.split():
        if word in user_greetings:
            return random.choice(bot_greetings)
        
```

```python
def index_sort(list_var):
    length=len(list_var)
    list_index=list(range(0,length))
    
    x= list_var
    for i in range(length):
        for j in range(length):
            if x[list_index[i]]>x[list_index[j]]:
                temp=list_index[i]
                list_index[i]=list_index[j]
                list_index[j]=temp
    return list_index
                    
```

```python
def bot_response(user_input):
    user_input=user_input.lower()
    sentence_list.append(user_input)
    bot_response=''
    cm=CountVectorizer().fit_transform(sentence_list)
    similarity_scores=cosine_similarity(cm[-1],cm)
    similarity_scores_list=similarity_scores.flatten()
    index=index_sort(similarity_scores_list)
    index=index[1:]
    print(index)
    response_flag=0
    
    j=0
    
    for i in range(len(index)):
        if similarity_scores_list[index[i]]>0:
            print(sentence_list[index[i]])
            bot_response= bot_response+' '+sentence_list[index[i]]
            print(bot_response)
            response_flag=1
            j=j+1
        if j>2:
            break
    if response_flag==0:
        bot_response=bot_response+' '+'I appolize, I dont understand'
    sentence_list.remove(user_input)
    return bot_response
```

```python
print("Tech Bot: I am Tech Bot for short. I will anwser your queries about iNeuronCourses.If you want to exit type bye")
exit_list=['exit','see you later','bye','quit','break']
while(True):
    user_input=input()
    if user_input.lower() in exit_list:
        print("Tech Bot: chat with you later!")
        break
    else:
        if greenting_response(user_input)!=None:
            print("Tech Bot:"+greenting_response(user_input))
        else:
            print('Tech Bot:'+bot_response(user_input))
    
```
